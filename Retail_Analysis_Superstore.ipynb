# CLEANING DATA

```python
df.isnull().sum()
#sum itu buat bikin true 1 false 0 agar tau setiap kolom punya berapa data null
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.lower()
```

```python
#hapus row yang ada null
df.dropna(inplace=True)
#isi dengan angka tertentu(0, median, mean) disaranin pake median
df['country'].fillna(0, inplace=True)
df['sales'].fillna(df['sales'].median(), inplace=True)
df['sales'].fillna(df['sales'].mean(), inplace=True)
```

IQR METHOD (Interquartile Range) Outlier Removal

```python
df.columns = df.columns.str.replace(' ','_').str.lower()
```

```python
df.info() #order_id harusnya bukan object jadi banyak yang salah
```

```python
#konversi data type date
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])
```

```python
df.info()
```

# ANALISIS DATA

```python
df.groupby('category')['sales'].sum().sort_values(ascending=False)
```

```python
df.groupby('category')['profit'].sum().sort_values(ascending=False) #furniture buat rugi sepertinya
```

```python
#remove outlier with IQR
quartile1 = df['sales'].quantile(0.25)
quartile3 = df['sales'].quantile(0.75)

iqr = quartile3 - quartile1

lower_bound = quartile1 - (1.5 * iqr)
upper_bound = quartile3 - (1.5 * iqr)

df = df[(df['sales'] >= lower_bound) & (df['sales'] <= upper_bound)]
```
